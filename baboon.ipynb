```python
import cv2
import numpy as np
from PIL import Image
im_path = "baboon.png"

im_array = cv2.imread(im_path)

display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```

```python
lower_black = np.array([0, 100, 100])
upper_black = np.array([9,255,255])
image_hsv = cv2.cvtColor(im_array, cv2.COLOR_BGR2HSV)

image_mask = cv2.inRange(image_hsv, lower_black, upper_black)
im_array[image_mask>0] = [0,0,0]
image_rgb = cv2.cvtColor(im_array, cv2.COLOR_HSV2BGR)


display(Image.fromarray(cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)))
```
